```python
import numpy as np
import pandas as pd
from pandas import Series
from pandas import DataFrame
```

```python
data=pd.read_csv("/content/drive/MyDrive/data_science files/bike_buyers.csv")
```

```python
# check whether the dataset has nan values
data.isnull().sum()
```

```python
# removing nan from "Marital Status" and "Children"
data[data["Children"].isnull()]
```

```python
data["Children"].value_counts()
```

```python
# if "Marital Status" is "Single" it means that that person does not have a child
# for married ones i am going to use "mode" command


for i in data[data["Children"].isnull()].index:
    if data.loc[i, "Marital Status"]=="Single":
        data.loc[i, "Children"]=0
    else:
        data.loc[i,"Children"]=2
data.isnull().sum()
# now we removed nans from "Children" column
```

```python
# Removing nans from "Marital Status"
# if the number of children is more than 0 it means that that perso is exactly married
# for the ones who don`t have children i am going write down "Single" But i know that it is not the correct way
# I wanted to remeve nans in other ways such as finding the correlation between "Age" and "Marital Status"
# I didn`t think off to implement this technique
data[data["Marital Status"].isnull()]
for i in data[data["Marital Status"].isnull()].index:
    if data.loc[i, "Children"]>0:
        data.loc[i, "Marital Status"]="Married"
    else:
        data.loc[i, "Marital Status"]="Single"
data.isnull().sum()
```

```python
# removing nans from "Gender" column
# to do this I am going to use "mode"
data["Gender"].value_counts()
data["Gender"].replace([np.nan],["Female"], inplace=True)
data.isnull().sum()
```

```python
# removing nans from "Income" column
# to do this I am going to use "median"
data["Income"].replace([np.nan], [data["Income"].median()], inplace=True)
```

```python
data["Income"].median()
```

```python
data.isnull().sum()
```

```python
# removing nans from "Home owner"
# if the income of that person is more than the median, I am going to write down "Yes"
# in the other case it is going to be "No"
# I know this is not the perfect solution
for i in data[data["Home Owner"].isnull()].index:
    if data.loc[i, "Income"]>60000:
       data.loc[i, "Home Owner"]="Yes"
    else:
       data.loc[i, "Home Owner"]="No"
```

```python
data.isnull().sum()
```

```python
# removing nans from "Cars"
data["Cars"].value_counts()
```

```python
# if the income is more than median i am going to write down 2 cars
# if the income is less than median i will write down 1 car
for i in data[data["Cars"].isnull()].index:
  if data.loc[i, "Income"]>60000: # the median
     data.loc[i, "Cars"]=2
  else:
     data.loc[i, "Cars"]=1
```

```python
data.isnull().sum()
```
